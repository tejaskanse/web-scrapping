```python
records=[]
```

```python
for res in results:
    date = res.find('strong').text[:-1] + ', 2017'
    lie = res.contents[1][1:-2] 
    exp = res.contents[2].find('a').text[1:-1] 
    url = res.find('a')['href'] 
 
    records.append([date, lie, exp, url])
```

```python
records
```

```python
import pandas as  pd
```

```python
df=pd.DataFrame(records,columns=["date","lie","explaination","url"])
```

```python
df["date"]=pd.to_datetime(df["date"])
```
